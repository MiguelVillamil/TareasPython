```python
import matplotlib.pyplot as plt

carros=[4976,4713,3890,3705,3653,3469]
etiquetas=['Mazda cx30','Renault Kwid','Chevrolet Onix','Renault Stepway','Renault Duster', 'Renault Sandero']
colores=['red','orange','blue','green','violet','yellow']
extraccion=[0.05,0.05,0.05,0.05,0.05,0.05]
plt.title('Carros mas vendidos en Colombia',fontsize = 20, color = "blue")
plt.pie(carros,labels=etiquetas,colors=colores,explode=extraccion,shadow=bool, autopct= "%1.2f%%")

plt.plot()
```

```python

caballos=[186,66,160,111,143,156]
precio=[108,44,52.7,50,74,48.8]

#hfont = {'fontname':'Helvetica'}
plt.title('Unidades vendidas \n VS \n Precio-Caballos de fuerza',fontsize = 20, color = "blue")#,**hfont)
plt.xlabel('Autos vendidos')
plt.ylabel("Caballos de Fuerza  Precio")
plt.plot(carros,caballos,color='red',linewidth=2,label='Caballos de Fuerza')
plt.plot(carros,precio,color='green',linewidth=2,label='Precio (Millones de COP)')
plt.legend()
plt.grid()

plt.plot()
```
